## Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Administrator\Documents\Pandas Project\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot()
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-deep')
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked =True, title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked =True, title = 'Flavor Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'barh', title = 'Ice Cream Ratings', stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Red')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize = (10,10))
```
